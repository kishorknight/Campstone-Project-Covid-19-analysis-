```python
lockdown_measures = ['school_closing', 'workplace_closing', 'cancel_events', 'gatherings_restrictions', 
                     'transport_closing', 'stay_home_restrictions', 'internal_movement_restrictions', 
                     'international_movement_restrictions']
plt.figure(figsize=(20, 20))
for i,measure in enumerate(lockdown_measures):
    plt.subplot(4, 2, i + 1)
    plt.plot(df['date'], df[measure], label=measure, color='r')
#     plt.plot(df['date'], df['daily_cases'], label='Daily Cases', color='b')
    plt.xlabel('Date')
    plt.ylabel('Value')
    plt.title(f"{measure} vs Daily Covid-19 Cases")
    plt.legend()
    plt.grid(True)
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
correlation_matrix = df[['daily_cases'] + lockdown_measures].corr()
print(correlation_matrix['daily_cases'].sort_values(ascending=False))
```

- Stronger Correlations (like workplace closing and stay-home restrictions) suggest that these measures are reactive to rising cases. Governments may enforce these restrictions when cases spike.
- Weaker Correlations (like transport or international restrictions) suggest that these measures may not have a significant direct impact on daily cases in India or may not be as critical once local transmission is widespread.

**Some lockdown measures may have lagging effect, where their impact is delayed by a few days or weeks.**
**For example, a lockdown measure today might only affect the number of cases after a few days (due to incubation periods, delays in transmission, etc.).**

```python
# Create lagged features for the lockdown measures
lag_days = [7, 14]  
data_lagged=df.copy()
for lag in lag_days:
    for measure in lockdown_measures:
        data_lagged[f'{measure}_lag{lag}'] = data_lagged[measure].shift(lag)

# Drop rows with NaN values generated by the shifting
data_lagged = data_lagged.dropna()

# Calculate correlations with daily cases for lagged features
for lag in lag_days:
    print(f'\nCorrelation with {lag}-day lag:')
    lagged_measures = [f'{measure}_lag{lag}' for measure in lockdown_measures]
    correlation_lagged = data_lagged[['daily_cases'] + lagged_measures].corr()
    print(correlation_lagged['daily_cases'].sort_values(ascending=False))
```

- **Stay Home Restrictions and Workplace Closings consistently have the strongest correlations with daily cases across both 7-day and 14-day lags. This suggests these measures are among the most effective in responding to rising cases, especially within the first week.**

- **The effect of School Closings and Internal Movement Restrictions is weaker but still observable, particularly after a 7-day lag.**

- **Transport and International Movement Restrictions do not show significant correlations, implying that their role in controlling daily cases is minimal within this dataset.**

# Performing statistical tests

## Independent t-test (Two-sample t-test)


```python
from scipy import stats

# T-test for daily cases when stay-home restrictions are applied vs not applied
stay_home_1 = df[df['stay_home_restrictions'] > 1]['daily_cases']
stay_home_0 = df[df['stay_home_restrictions'] <= 1]['daily_cases']

# Perform the t-test
t_stat, p_value = stats.ttest_ind(stay_home_1, stay_home_0)

print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

- Null Hypothesis: There is no significant difference in the daily cases between the days with stay-home restrictions applied and those without. This means that the mean daily cases for the two groups are equal.

- Alternative Hypothesis : There is a significant difference in the daily cases between the days with stay-home restrictions applied and those without. This means that the mean daily cases for the two groups are not equal.

- P-value: The p-value of approximately 3.88×10 −9 is extremely low. In hypothesis testing, a common threshold to reject the null hypothesis is 0.05 (5%). Since our p-value is much smaller than 0.05, we can reject the null hypothesis.

**There is strong evidence to conclude that there is a statistically significant difference in daily cases when stay-home restrictions are applied compared to when they are not applied. This suggests that stay-home restrictions may be associated with changes in the number of daily COVID-19 cases.**

```python
data=df.copy()
```

```python
data['stay_home_group'] = data['stay_home_restrictions'].apply(lambda x: 'No Restrictions' if x <= 1 else 'With Restrictions')
plt.figure(figsize=(10, 6))
sns.boxplot(x='stay_home_group', y='daily_cases', data=data)
plt.title('Daily Cases Based on Stay-Home Restrictions', fontsize=16)
plt.xlabel('Stay-Home Restrictions', fontsize=14)
plt.ylabel('Daily Cases', fontsize=14)
plt.show()
```

- This might indicate that stay-home restrictions were applied when the number of daily cases was already high, or that regions with higher cases were more likely to implement strict stay-home policies.
- It does not necessarily mean that stay-home restrictions cause an increase in cases. Instead, they may be a response to rising cases.

```python
data['work_home'] = data['workplace_closing'].apply(lambda x: 'No Restrictions' if x <= 1 else 'With Restrictions')
plt.figure(figsize=(10, 6))
sns.boxplot(x='work_home', y='daily_cases', data=data)
plt.title('Daily Cases Based on Working Restrictions', fontsize=16)
plt.xlabel('Work from Home Restrictions', fontsize=14)
plt.ylabel('Daily Cases', fontsize=14)
plt.show()
```

```python
data['school_group'] = data['school_closing'].apply(lambda x: 'No Restrictions' if x <= 1 else 'With Restrictions')
plt.figure(figsize=(10, 6))
sns.boxplot(x='school_group', y='daily_cases', data=data)
plt.title('Daily Cases Based on School Closing Restrictions', fontsize=16)
plt.xlabel('School Closing Restrictions', fontsize=14)
plt.ylabel('Daily Cases', fontsize=14)
plt.show()
```

## ANOVA (Analysis of Variance)

```python
# Group daily cases based on different levels of workplace_closing
group_0 = df[df['workplace_closing'] == 0]['daily_cases']
group_1 = df[df['workplace_closing'] == 1]['daily_cases']
group_2 = df[df['workplace_closing'] == 2]['daily_cases']
group_3 = df[df['workplace_closing'] == 3]['daily_cases']

# Perform ANOVA test
f_stat, p_value = stats.f_oneway(group_0, group_1, group_2, group_3)

print(f"F-statistic: {f_stat}, P-value: {p_value}")
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='workplace_closing', y='daily_cases', data=df)
plt.title('Daily Cases vs. Workplace Closing Levels')
plt.xlabel('Workplace Closing Levels')
plt.ylabel('Daily Cases')
plt.show()
```

```python
# Group daily cases based on different levels of transport_closing
group_0 = df[df['transport_closing'] == 0]['daily_cases']
group_1 = df[df['transport_closing'] == 1]['daily_cases']
group_2 = df[df['transport_closing'] == 2]['daily_cases']

# Perform ANOVA test
f_stat, p_value = stats.f_oneway(group_0, group_1, group_2)

print(f"F-statistic: {f_stat}, P-value: {p_value}")

```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='transport_closing', y='daily_cases', data=df)
plt.title('Daily Cases vs. Transport Closing Levels')
plt.xlabel('Transport Closing Levels')
plt.ylabel('Daily Cases')
plt.show()
```

```python
# ANOVA test for school_closing
school_closing_0 = df[df['school_closing'] == 0]['daily_cases']
school_closing_1 = df[df['school_closing'] == 1]['daily_cases']
school_closing_2 = df[df['school_closing'] == 2]['daily_cases']
school_closing_3 = df[df['school_closing'] == 3]['daily_cases']

# Perform ANOVA test
f_stat, p_value = stats.f_oneway(school_closing_0, school_closing_1, school_closing_2, school_closing_3)

print(f"F-statistic: {f_stat}, P-value: {p_value}")
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='school_closing', y='daily_cases', data=df)
plt.title('Daily Cases vs. School Closing Levels')
plt.xlabel('School Closing Levels')
plt.ylabel('Daily Cases')
plt.show()
```

```python
# ANOVA for Internal Movement Restrictions
internal_movement_0 = df[df['internal_movement_restrictions'] == 0]['daily_cases']
internal_movement_1 = df[df['internal_movement_restrictions'] == 1]['daily_cases']
internal_movement_2 = df[df['internal_movement_restrictions'] == 2]['daily_cases']

f_stat_internal, p_value_internal = stats.f_oneway(internal_movement_0, internal_movement_1, internal_movement_2)
print(f"Internal Movement Restrictions ANOVA: F-statistic: {f_stat_internal}, P-value: {p_value_internal}")

# ANOVA for Cancel Events
cancel_events_0 = df[df['cancel_events'] == 0]['daily_cases']
cancel_events_1 = df[df['cancel_events'] == 1]['daily_cases']
cancel_events_2 = df[df['cancel_events'] == 2]['daily_cases']

f_stat_cancel, p_value_cancel = stats.f_oneway(cancel_events_0, cancel_events_1, cancel_events_2)
print(f"Cancel Events ANOVA: F-statistic: {f_stat_cancel}, P-value: {p_value_cancel}")

# ANOVA for International Movement Restrictions
international_movement_0 = df[df['international_movement_restrictions'] == 0]['daily_cases']
international_movement_1 = df[df['international_movement_restrictions'] == 1]['daily_cases']
international_movement_2 = df[df['international_movement_restrictions'] == 2]['daily_cases']
international_movement_3 = df[df['international_movement_restrictions'] == 3]['daily_cases']
international_movement_4 = df[df['international_movement_restrictions'] == 4]['daily_cases']

f_stat_international, p_value_international = stats.f_oneway(international_movement_0, international_movement_1, international_movement_2, international_movement_3, international_movement_4)
print(f"International Movement Restrictions ANOVA: F-statistic: {f_stat_international}, P-value: {p_value_international}")
```

## Critical Relationships and Hidden Insights
